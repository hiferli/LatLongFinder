```python
import pgeocode
import pandas as pd
```

```python
# Function to get Latitude and Longitude from the Pincode and Country
def pinToLatLong(postalcode, country='in'):
    nomi = pgeocode.Nominatim(country)
    location = nomi.query_postal_code(postalcode)
    return location.latitude, location.longitude
```

```python
# Reading Data from CSV File
df = pd.read_csv('/content/Pin_code.csv')
```

```python
print(df)
```

```python
# Making new columns for Lat and Long for every Pincode in the Dataframe
df[['Latitude', 'Longitude']] = df['PIN Code'].apply(pinToLatLong).apply(pd.Series)
```
